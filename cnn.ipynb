```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import numpy as np
```

```python
train_datagen=ImageDataGenerator(
    rescale=1/255,
    shear_range=.2,
    zoom_range=.2,
    horizontal_flip=True)
training_set=train_datagen.flow_from_directory(
r"C:\Users\jrfid\OneDrive\Documents\python\dataset\dataset\training_set",    
    target_size=(64,64),
    batch_size=32,
    class_mode='binary')
    
```

```python
test_datagen=ImageDataGenerator(rescale=1/255)
test_set=test_datagen.flow_from_directory(
    r"C:\Users\jrfid\OneDrive\Documents\python\dataset\dataset\test_set",
    target_size=(64,64),
    batch_size=32,
    class_mode='binary')
    
```

```python
#cnn
cnn=tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu',input_shape=[64,64,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x=training_set,validation_data=test_set,epochs=10)
```

```python
from keras.preprocessing import image
test_image=image.load_img(r"C:\Users\jrfid\Downloads\255883.jpg",target_size=(64,64))
test_image=image.img_to_array(test_image)
test_image=np.expand_dims(test_image,axis=0)
result=cnn.predict(test_image)
training_set.class_indices
if result[0][0]==1:
    prediction='dog'
else:
    prediction='cat'

print(prediction)


```
